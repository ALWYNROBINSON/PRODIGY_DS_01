```python
for i in columns:
    fig=plt.figure(figsize=(5,5))
    plt.hist(d[i],color='#B22222',bins=10)
    plt.xlabel(i)
    plt.show()
```

```python
yrs=d.columns[1:]
tot=d[yrs].sum()
plt.figure(figsize=(30,30))
plt.barh(yrs,tot,color='#191970')
plt.xlabel('Total Values')
plt.ylabel('Years',size=20)
plt.title('Total Values per year',size=20)
plt.show()
```

```python
c1960=d.sort_values(by='1960').head(20)
```

```python
c1960
```

```python
c1960_1=c1960.set_index('Country Name')
for cn,dv in c1960_1.iterrows():
    fig=plt.figure(figsize=(10,5))
    sns.barplot(x=dv.index,y=dv.values)
    plt.ylabel('Data Values')
    plt.xlabel('Years')
    plt.title(f"{cn} - Data Values from 1960 to 2022")
    plt.xticks(rotation=90)
    plt.show()
```
